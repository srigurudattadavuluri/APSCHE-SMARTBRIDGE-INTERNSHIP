## GENERALLY I BUILD THE BASE MODEL WITHOUT BALAACING THE CLASSES BUT IT IS NECESSARY TO BALACE THE CLASSES AS IF YOU SEE

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.3)
```

```python
x_train
```

```python
print("Before Balancing:")
print(y_train.value_counts())
print(y_train.value_counts(normalize=True))
```

```python
train_data = x_train.copy()
train_data['STATUS'] = y_train
```

```python
from sklearn.utils import resample
import pandas as pd

majority_class = train_data[train_data['STATUS'] == 0]
minority_class = train_data[train_data['STATUS'] == 1]

desired_count = len(train_data) // 2

majority_resampled = resample(
    majority_class,
    replace=False,
    n_samples=desired_count,
    random_state=42
)

minority_resampled = resample(
    minority_class,
    replace=True,
    n_samples=desired_count,
    random_state=42
)

train_balanced = pd.concat(
    [majority_resampled, minority_resampled]
).sample(frac=1, random_state=42)
```

```python
train_balanced
```

```python
x_train_balanced = train_balanced.drop('STATUS', axis=1)
y_train_balanced = train_balanced['STATUS']
```
